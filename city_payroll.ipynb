```python
# creating variables to get number of null values for employee names
null_first_last = df[(df['last_name'].isnull()) & (df['first_name'].isnull())].shape[0]
null_last = df[df['last_name'].isnull()].shape[0]
null_first = df[df['first_name'].isnull()].shape[0]
null_middle = df[df['mid_init'].isnull()].shape[0]
num_entries = 6225611
```

```python
# filling null names and initials with 'unknown'
df[['last_name', 'first_name', 'mid_init']] = df[['last_name', 'first_name', 'mid_init']].fillna('unknown')
```
